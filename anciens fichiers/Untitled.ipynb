```python
def verif_prod(imdb_id):
    
    
    #SELECT PRODUCTION
    
    r4 = requests.get("https://www.imdb.com/title/"+imdb_id+"/companycredits")
    bs4 = BeautifulSoup(r4.content)
    
    prod = bs4.find_all('h4',{"id":"production"})

    return len(prod)
```

```python
test_dict = {}
```

```python
for i in tqdm.tqdm(movie_no_1):
    if verif_prod(i) > 0:
        test_dict[i] = 1
    else:
        test_dict[i] = 0
```

```python
def skip_prod(x):
    for i in range(len(list(test_dict.keys()))):
        if x == list(test_dict.keys())[i] and list(test_dict.values())[i] == 0:
            result = ""
        else :
            result = x
    return result
```
