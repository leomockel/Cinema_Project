```python
actor_1 = df["Actor 1"].values
actor_2 = df["Actor 2"].values
actor_3 = df["Actor 3"].values
actor_4 = df["Actor 4"].values
actor_5 = df["Actor 5"].values
actor_6 = df["Actor 6"].values
actor_7 = df["Actor 7"].values
actor_8 = df["Actor 8"].values
actor_9 = df["Actor 9"].values
actor_10 = df["Actor 10"].values
actor_11 = df["Actor 11"].values
actor_12 = df["Actor 12"].values
actor_13 = df["Actor 13"].values
actor_14 = df["Actor 14"].values
actor_15 = df["Actor 15"].values
actor_16 = df["Actor 16"].values
actor_17 = df["Actor 17"].values
actor_18 = df["Actor 18"].values
actor_19 = df["Actor 19"].values
actor_20 = df["Actor 20"].values
actor_21 = df["Actor 21"].values
actor_22 = df["Actor 22"].values
actor_23 = df["Actor 23"].values
actor_24 = df["Actor 24"].values
actor_25 = df["Actor 25"].values
actor_26 = df["Actor 26"].values
actor_27 = df["Actor 27"].values
actor_28 = df["Actor 28"].values
actor_29 = df["Actor 29"].values
```

```python
top_actors = list(actor_1)+list(actor_2)+list(actor_3)+list(actor_4)+list(actor_5)[:37235]
```

```python
dict_top_act = {}
```

```python
for name in top_actors:
    dict_top_act[name] = dict_top_act.get(name, 0) + 1
```

```python
len(dict_top_act.keys())
```

```python
sec_actors = list(actor_5)[37235:]+list(actor_6)+list(actor_7)+list(actor_8)+list(actor_9)+list(actor_10)[:33111]
```

```python
dict_sec_act = {}
```

```python
for name in sec_actors:
    if name in dict_top_act.keys():
        dict_top_act[name] += 1
    else:
        dict_sec_act[name] = dict_sec_act.get(name, 0) + 1
```

```python
len(dict_sec_act.keys())
```

```python
third_actors = list(actor_10)[33111:]+list(actor_11)+list(actor_12)+list(actor_13)+list(actor_14)+list(actor_15)[:56737]
```

```python
dict_third_act = {}
```

```python
for name in third_actors:
    if name in dict_top_act.keys():
        dict_top_act[name] += 1
    elif name in dict_sec_act.keys():
        dict_sec_act[name] += 1
    else:
        dict_third_act[name] = dict_third_act.get(name, 0) + 1
```

```python
len(dict_third_act.keys())
```

```python
rare_actors = list(actor_15)[56737:]+list(actor_16)+list(actor_17)+list(actor_18)+list(actor_19)+list(actor_20)+list(actor_21)[:71845]
```

```python
dict_rare_act = {}
```

```python
for name in rare_actors:
    if name in dict_top_act.keys():
        dict_top_act[name] += 1
    elif name in dict_sec_act.keys():
        dict_sec_act[name] += 1
    elif name in dict_third_act.keys():
        dict_third_act[name] += 1    
    else:
        dict_rare_act[name] = dict_rare_act.get(name, 0) + 1
```

```python
len(dict_rare_act.keys())
```

```python
v_rare_actors = list(actor_21)[71845:]+list(actor_22)+list(actor_23)+list(actor_24)+list(actor_25)+list(actor_26)+list(actor_27)+list(actor_28)+list(actor_29)
```

```python
dict_vrare_act = {}
```

```python
for name in v_rare_actors:
    if name in dict_top_act.keys():
        dict_top_act[name] += 1
    elif name in dict_sec_act.keys():
        dict_sec_act[name] += 1
    elif name in dict_third_act.keys():
        dict_third_act[name] += 1   
    elif name in dict_rare_act.keys():
        dict_rare_act[name] += 1   
    else:
        dict_vrare_act[name] = dict_vrare_act.get(name, 0) + 1
```

```python
len(dict_vrare_act.keys())
```

```python
col_names = ["top_actors", "top_act_num", "sec_actors", "sec_act_num", "third_actors", "third_act_num", 
             "rare_actors", "rare_act_num", "v_rare_actors", "vrare_act_num"]
```

```python
new_df = pd.DataFrame(columns=col_names)
```

```python
new_df["top_actors"] = dict_top_act.keys()
new_df["top_act_num"] = dict_top_act.values()
new_df["sec_actors"] = dict_sec_act.keys()
new_df["sec_act_num"] = dict_sec_act.values()
new_df["third_actors"] = dict_third_act.keys()
new_df["third_act_num"] = dict_third_act.values()
new_df["rare_actors"] = dict_rare_act.keys()
new_df["rare_act_num"] = dict_rare_act.values()
new_df["v_rare_actors"] = dict_vrare_act.keys()
new_df["vrare_act_num"] = dict_vrare_act.values()
```

### Find actors without codes

###### Get lists of actors

```python
top_act = new_df.top_actors.values.tolist()
```

```python
sec_act = new_df.sec_actors.values.tolist()
```

```python
thi_act = new_df.third_actors.values.tolist()
```

```python
rar_act = new_df.rare_actors.values.tolist()
```

```python
vra_act = new_df.v_rare_actors.values.tolist()
```

```python
actor_find = []
```

###### Find actors without code, by comparing with code dictionary values

```python
for name in tqdm.tqdm(top_act):
    if name not in code_dict.values():
        actor_find.append(name)
```

```python
for name in tqdm.tqdm(sec_act):
    if name not in code_dict.values():
        actor_find.append(name)
```

```python
for name in tqdm.tqdm(thi_act):
    if name not in code_dict.values():
        actor_find.append(name)
```

```python
for name in tqdm.tqdm(rar_act):
    if name not in code_dict.values():
        actor_find.append(name)
```

```python
for name in tqdm.tqdm(vra_act):
    if name not in code_dict.values():
        actor_find.append(name)
```
